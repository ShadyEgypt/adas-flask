```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Names of your CSV files
files = ['0x01-yolov8-graph.csv','0x02-yolov8-opt-graph.csv','0x03-adas1-graph.csv','0x04-adas2-graph.csv']
```

```python
labels = ['YOLOv8-pytorch','YOLOv8-ONNX','Open Zoo road segmentation','Open Zoo semantic segmentation']
```

```python
# Read each CSV file and convert the inference time to milliseconds
dataframes = []
for file in files:
    df = pd.read_csv(file)
    df['Inference Time'] *= 1000  # Assuming the original time is in seconds
    dataframes.append(df)
```

```python
plt.figure(figsize=(10, 6))
for i, df in enumerate(dataframes):
    plt.plot(df['index'], df['Inference Time'], label=labels[i])

plt.xlabel('Index')
plt.ylabel('Inference Time (milliseconds)')
plt.title('Inference Time Comparison')
plt.legend()
plt.show()
```
